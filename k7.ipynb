```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

image = cv2.imread("image444.png")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
channels = image.shape[2]
print('Количество каналов : ',channels)
```

```python
#проверяем наличие прозрачных пикселей
from PIL import Image  
def has_transparency(image_path):  
    try:  
        return "A" in Image.open(image_path).getbands()  
    except Exception:  
        return False  

result = has_transparency("image33.png") 

print("Изображение имеет прозрачные пиксели." if result else "Изображение не имеет прозрачных пикселей.")
```

```python
# здесь заранее подготовленное изображение с прозрачным фоном
import cv2
import matplotlib.pyplot as plt
import numpy as np

image = cv2.imread("cat.png")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
channels = image.shape[2]
print('Количество каналов : ',channels)
```

```python
#проверяем наличие прозрачных пикселей
from PIL import Image  
def has_transparency(image_path):  
    try:  
        return "A" in Image.open(image_path).getbands()  
    except Exception:  
        return False  

result = has_transparency("cat.png") 

print("Изображение имеет прозрачные пиксели." if result else "Изображение не имеет прозрачных пикселей.")
```

```python
#заменяем прозрачные пиксели белыми непрозрачными
import cv2
import matplotlib.pyplot as plt
import numpy as np
image = cv2.imread("cat.png", cv2.IMREAD_UNCHANGED)
alpha_channel = image[:, :, 3]
_, mask = cv2.threshold(alpha_channel, 254, 255, cv2.THRESH_BINARY)
color = image[:, :, :3]
new_image = cv2.bitwise_not(cv2.bitwise_not(color, mask=mask))
plt.imshow(new_image)
```

### Задание №2. Реализовать функции для изменения размера изображения, конвертации в черно-белое и применить фильтр Гаусса (размытие).

```python
import cv2
import numpy as np
image = cv2.imread('image444.png')
cv2.imshow('image444.png', image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#изменение размера картинки с учетом соотношения сторон
import cv2
import numpy as np
image = cv2.imread('image444.png')
cv2.imshow('image444.png', image)
scale_percent = 50
width = int(image.shape[1] * scale_percent / 100)
height = int(image.shape[0] * scale_percent / 100)
dim = (width, height)
resized = cv2.resize(image, dim, interpolation = cv2.INTER_AREA)
cv2.imshow('image444.png', resized)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#использование фильтра Гаусса
import cv2 
import numpy as np 
  
image = cv2.imread('image444.png') 
  
cv2.imshow('image444.png', image) 
cv2.waitKey(0) 
  
Gaussian = cv2.GaussianBlur(image, (7, 7), 0) 

cv2.imshow('Gaussian Blurring', Gaussian) 
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#преобразование картинки в серую
import cv2

img = cv2.imread('image444.png')
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

cv2.imshow('image444.png', gray_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#преобразование картинки в чернобелую
import cv2

im_gray = cv2.imread('image444.png', cv2.IMREAD_GRAYSCALE)
(thresh, im_bw) = cv2.threshold(im_gray, 128, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
thresh = 127
im_bw = cv2.threshold(im_gray, thresh, 255, cv2.THRESH_BINARY)[1]

cv2.imshow('image444.png', im_bw)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#еще один вариант преобразования картинки в чернобелую
import cv2
color_image = cv2.imread('image444.png')

gray_image = cv2.cvtColor(color_image, cv2.COLOR_BGR2GRAY)
_, binary_image = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)

cv2.imshow('Binary Image', binary_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
